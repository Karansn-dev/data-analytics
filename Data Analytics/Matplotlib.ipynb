```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6]
y = [6,7,8,9,10,11]
plt.plot(x,y)
plt.title("simple plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6]
y = [6,7,8,9,10,11]
plt.plot(x,y , color = 'red', linestyle = 'dashed', marker = 'o', markerfacecolor = 'blue', markersize = 10)
plt.subplot(2,2,1)
plt.title("simple plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]

plt.subplot(2,2,1)
plt.plot(x, [1,4,9,16])
plt.title("Plot 1")

plt.subplot(2,2,2)
plt.plot(x, [1,2,3,4])
plt.title("Plot 2")

plt.subplot(2,2,3)
plt.plot(x, [4,3,2,1])
plt.title("Plot 3")

plt.subplot(2,2,4)
plt.plot(x, [16,9,4,1])
plt.title("Plot 4")

plt.tight_layout()
plt.show()
```
